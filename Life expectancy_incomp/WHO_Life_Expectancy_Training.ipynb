We need to impute a lot of columns in the given dataset. We can first impute Life expectancy, Adult Mortality, Polio, Diphtheria, thinness attributes, BMI, Schooling, Income composition of resources, Alcohol, total expenditure, GDP, Hep B, Population in the same order.

We can directly drop tables like under-five deaths, GDP, thinness 1-19 due to presence of immideate substitutes.

```python
df.drop(columns=['under-five deaths ', 'GDP', ' thinness 1-19 years'], inplace=True, axis=1)
sns.heatmap(df.corr(), annot=True)
```
